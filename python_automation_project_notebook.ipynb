## 1) Import Libraries

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

## 2) Read the Dataset

```python
df = pd.read_csv("python_automation_sales_orders.csv")
df.head()
```

## 3) First Look at the Data
نبدأ نفهم:
- عدد الصفوف والأعمدة
- أسماء الأعمدة
- أنواع البيانات
- Missing values

```python
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())

print("\nInfo:")
df.info()

print("\nMissing Values:")
print(df.isna().sum().sort_values(ascending=False))
```

```python
df.describe()
```

## 4) Convert Data Types
- `order_date` → datetime
- الأعمدة الرقمية → numeric

```python
df["order_date"] = pd.to_datetime(df["order_date"], format="%Y-%m-%d", errors="coerce")

numeric_cols = ["quantity", "discount_pct", "delivery_days", "customer_rating", "unit_price"]
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

df.info()
```

## 5) Text Cleaning
هننضف الأعمدة النصية من:
- المسافات الزائدة
- اختلاف الـ case

```python
def clean_text(value):
    if pd.isna(value):
        return value
    return str(value).strip().title()

text_cols = ["payment_method", "customer_city", "product_category", "customer_segment", "product_name"]

for col in text_cols:
    df[col] = df[col].apply(clean_text)

df[text_cols].head()
```

## 6) Remove Duplicates

```python
df.duplicated().sum()
```

```python
before = len(df)
df = df.drop_duplicates()
#==== df.drop_duplicates(inplace=True) --- IGNORE ---
after = len(df)

print("Duplicates removed:", before - after)
print("Shape after removing duplicates:", df.shape)
```

## 7) Handle Missing Values
### Business Logic:
- `discount_pct` missing → 0
- `customer_city` missing → `"Unknown"`
- `customer_rating` missing → median
- `unit_price` missing → median by category, then overall median

```python
df.isnull().sum()
```

```python
df["discount_pct"] = df["discount_pct"].fillna(0)
df["customer_city"] = df["customer_city"].fillna("Unknown")
df["customer_rating"] = df["customer_rating"].fillna(df["customer_rating"].median())

df["unit_price"] = df.groupby("product_category")["unit_price"].transform(lambda s: s.fillna(s.median()))
#df["unit_price"] = df["unit_price"].fillna(df["unit_price"].median())

df.isna().sum().sort_values(ascending=False)
```

## 8) Handle Invalid Values
### Rules:
- `quantity <= 0` تعتبر invalid
- `delivery_days < 0` تعتبر invalid

```python
df.loc[df["quantity"] <= 0, "quantity"] = np.nan
df.loc[df["delivery_days"] < 0, "delivery_days"] = np.nan

df.isnull().sum().sort_values(ascending=False)  
```

```python
df.dropna(subset=["quantity", "delivery_days"], inplace=True)

df.isnull().sum().sort_values(ascending=False)  
```

## 9) Feature Engineering
هنضيف:
- `month`
- `gross_revenue`
- `net_revenue`

```python
df["month"] = df["order_date"].dt.month_name()
df["gross_revenue"] = df["unit_price"] * df["quantity"]
df["net_revenue"] = df["gross_revenue"] * (1 - df["discount_pct"] / 100)

df[["order_date", "month", "unit_price", "quantity", "gross_revenue", "net_revenue"]].head()
```

```python
df.to_csv("python_automation_sales_orders_cleaned.csv", index=False)
```

## 10) KPI Summary

```python
total_orders = df["order_id"].nunique()
total_customers = df["customer_id"].nunique()
total_revenue = round(df["net_revenue"].sum(), 2)
avg_order_value = round(df.groupby("order_id")["net_revenue"].sum().mean(), 2)
return_rate = round(df["returned_flag"].mean() * 100, 2)
avg_delivery_days = round(df["delivery_days"].mean(), 2)

kpi_summary = pd.DataFrame({
    "metric": [
        "total_orders",
        "total_customers",
        "total_net_revenue",
        "average_order_value",
        "return_rate_pct",
        "avg_delivery_days"
    ],
    "value": [
        total_orders,
        total_customers,
        total_revenue,
        avg_order_value,
        return_rate,
        avg_delivery_days
    ]
})

kpi_summary
```

## 11) Revenue by Category

```python
revenue_by_category = (
    df.groupby("product_category", as_index=False)
      .agg(
          total_net_revenue=("net_revenue", "sum"),
          total_quantity=("quantity", "sum"),
          avg_rating=("customer_rating", "mean")
      )
      .sort_values("total_net_revenue", ascending=False)
)

revenue_by_category.head(10)
```

## 12) Revenue by City

```python
revenue_by_city = (
    df.groupby("customer_city", as_index=False)
      .agg(
          total_net_revenue=("net_revenue", "sum"),
          total_orders=("order_id", "nunique"),
          avg_delivery_days=("delivery_days", "mean")
      )
      .sort_values("total_net_revenue", ascending=False)
)

revenue_by_city.head(10)
```

## 13) Monthly Sales Report

```python
monthly_sales = (
    df.groupby("month", as_index=False)
      .agg(
          total_net_revenue=("net_revenue", "sum"),
          total_orders=("order_id", "nunique"),
          return_rate=("returned_flag", "mean")
      )
      .sort_values("month")
)

monthly_sales["return_rate_pct"] = (monthly_sales["return_rate"] * 100).round(2)

monthly_sales
```

## 14) Returned Orders Report

```python
returned_orders = df[df["returned_flag"] == 1].copy()
returned_orders = returned_orders[[
    "order_id", "order_date", "customer_id", "customer_city",
    "product_category", "product_name", "net_revenue", "delivery_days"
]].sort_values(["order_date", "order_id"])

returned_orders.head(10)
```

## 15) Top Products

```python
top_products = (
    df.groupby(["product_id", "product_name", "product_category"], as_index=False)
      .agg(
          total_net_revenue=("net_revenue", "sum"),
          total_quantity=("quantity", "sum")
      )
      .sort_values("total_net_revenue", ascending=False)
      .head(15)
)

top_products
```

## 16) Export Reports to CSV

```python
output_path = Path("automation_reports")
output_path.mkdir(exist_ok=True)

df.to_csv(output_path / "cleaned_sales_orders.csv", index=False)
kpi_summary.to_csv(output_path / "kpi_summary.csv", index=False)
revenue_by_category.to_csv(output_path / "revenue_by_category.csv", index=False)
revenue_by_city.to_csv(output_path / "revenue_by_city.csv", index=False)
monthly_sales.to_csv(output_path / "monthly_sales.csv", index=False)
returned_orders.to_csv(output_path / "returned_orders.csv", index=False)
top_products.to_csv(output_path / "top_products.csv", index=False)

print("CSV reports exported successfully.")
```

## 17) Export Reports to Excel (Multiple Sheets)

```python
with pd.ExcelWriter(output_path / "sales_report.xlsx", engine="openpyxl") as writer:
    kpi_summary.to_excel(writer, sheet_name="KPI Summary", index=False)
    revenue_by_category.to_excel(writer, sheet_name="By Category", index=False)
    revenue_by_city.to_excel(writer, sheet_name="By City", index=False)
    monthly_sales.to_excel(writer, sheet_name="Monthly Sales", index=False)
    top_products.to_excel(writer, sheet_name="Top Products", index=False)
    returned_orders.to_excel(writer, sheet_name="Returned Orders", index=False)

print("Excel report exported successfully.")
```
